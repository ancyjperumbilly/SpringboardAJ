```python
inter_mth= inter_mth.drop(columns=['Win Conditions'])
inter_mth.head()
```

```python
wld_mth = wld_mth.drop(columns=['Win Conditions'])
wld_mth.head()
```

```python
inter_mth['Year']=pd.DatetimeIndex(inter_mth['Date']).year
```

```python
grp=fifa_rank.groupby('Group')
Fifa_grp_2022=list(fifa_rank['Team'].unique())
Fifa_grp_2022
```

```python
My_df=[]
for j in Fifa_grp_2022:
    win_inter =len(inter_mth[(inter_mth['Winning Team']==j) & (inter_mth['Year']>1998)])
    loss_inter=len(inter_mth[(inter_mth['Losing Team']==j) & (inter_mth['Year']>1998)])
    tot_inter=win_inter+loss_inter
    if tot_inter==0:
        win_inter_perct=0
        loss_inter_perct=0
    else:
        win_inter_perct=(win_inter/tot_inter)*100
        loss_inter_perct=(loss_inter/tot_inter)*100

    win_wrld =len(wld_mth[(wld_mth['Winning Team']==j) & (wld_mth['Year']>1998)])
    loss_wrld=len(wld_mth[(wld_mth['Losing Team']==j) & (wld_mth['Year']>1998)])
    tot_wrld=win_wrld+loss_wrld
    if tot_wrld==0:
        win_wrld_perct=0
        loss_wrld_perct=0
    else:
        win_wrld_perct=(win_wrld/tot_wrld)*100
        loss_wrld_perct=(loss_wrld/tot_wrld)*100
    
    team_rnk=list(fifa_rank[fifa_rank['Team']==j]['FIFA Ranking'])[0]
    dic1={'Country':j}
    grp_df=pd.DataFrame([[j,win_inter_perct,loss_inter_perct,win_wrld_perct,loss_wrld_perct,team_rnk]],columns=['Country','International_Mth_Win','International_Mth_Loss','Wrld_Cup_Win','Wrld_Cup_Loss','Fifa2022_Rnk'])
    My_df.append(grp_df)
    

Fin_Dataset=pd.concat(My_df)
```

```python
Fin_Dataset
```

```python
Fin_Dataset.sort_values(by=['Fifa2022_Rnk'])
```

```python
Fin_Dataset['Inter_Win_rank']=Fin_Dataset['International_Mth_Win'].rank(ascending=False)
```

```python
Fin_Dataset['WrldCup_Win_Rnk']=Fin_Dataset['Wrld_Cup_Win'].rank(ascending=False)
```

```python
Fin_Dataset['My_2022_FiFa_Prediction']=(Fin_Dataset['Inter_Win_rank']*0.30+Fin_Dataset['WrldCup_Win_Rnk']*0.70)/2
```

```python
Fin_Dataset['My_2022_FiFa_Prediction']=Fin_Dataset['My_2022_FiFa_Prediction'].apply(lambda w:math.floor(w))
```
